### 1) Leer y representar la imagen riceBGgradient.tif

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

from PIL import Image

from skimage.morphology import (square, rectangle, diamond, disk, cube,
                                octahedron, ball, octagon, star,
                                binary_erosion, binary_dilation, binary_opening, binary_closing,
                               remove_small_holes, label, convex_hull_image, skeletonize, thin, reconstruction,
                               opening, closing, dilation, erosion)

from skimage import data
from skimage.util import invert
from skimage import data, util, exposure, color
from scipy.ndimage import convolve, correlate, uniform_filter, gaussian_filter, median_filter, rotate 
from scipy.ndimage import prewitt, sobel, laplace

from skimage import filters
from skimage import feature
```

```python
file = 'images/riceBGgradient.tif'
img  = Image.open(file)

print('formato:', img.format)
print('(ancho, alto) =', img.size)
print('modo: ', img.mode)

img = np.array(img)

plt.imshow(img, cmap='gray')
```

### 2) Intenta separar los granos del fondo estableciendo un umbral

Primero visualizamos el histograma para poder tener un criterio a la hora de seleccionar el umbral

```python
plt.figure(figsize=(8, 4))

img_hist, bins = exposure.histogram(img, normalize=False)
plt.plot(bins, img_hist)
plt.axis([0,255, 0, 1.01*img_hist.max()])
plt.title('Histograma')
plt.xlabel('Valor (intensidad)')
plt.ylabel('Frecuencias absolutas')
plt.grid()
```

```python
imagenGranos = np.copy(img)
```

```python
th = 135

mask = imagenGranos < th
imagenGranos[mask] = 0
plt.imshow(imagenGranos, cmap='gray')
```

Hemos ido probando valores para el umbral, hasta llegar a 130, donde vemos que, todav√≠a no se ha eliminado el fondo del todo y ya se empiezan a borrar algunos granos. Por tanto, podemos concluir que con este m√©todo tan simple, no podemos aislar los granos, necesitariamos algo m√°s.

### 3) Como habr√°s visto, no es posible obtener un resultado satisfactorio debido a la inhomogeneidad en la iluminaci√≥n. Utiliza una transformaci√≥n morfol√≥gica top-hat para corregir la iluminaci√≥n, tal y como se ha hecho en clase

Como vemos que el fondo tiene distintas intesidades dependiendo de la parte de la imagen en la que nos fijemos, vamos a intentar igualar el fondo para poder eliminarlo homog√©neamente. Para ello vamos a aplicar una transformaci√≥n Top-Hat.

```python
fondo = opening(img, disk(40))
plt.imshow(fondo, cmap='gray')
```

```python
imgHomo = img-fondo

plt.figure(figsize = (12,8))

plt.subplot(121)
plt.imshow(img, cmap='gray')
plt.title('Imagen original')

plt.subplot(122)
plt.imshow(imgHomo, cmap='gray')
plt.title('Imagen fondo homog√©neo')
```

Como podemos ver, ahora el fondo es mucho m√°s homog√©neo en todas la imagen

### 4) Busca ahora un umbral. Para buscar el umbral, puedes ayudarte con la funci√≥n ‚Äúskimage.filters.threshold_otsu‚Äù

Ahora con la imagen del fondo homog√©neo, vamos a aplicar el umbral para separar los granos del fondo aplicando una funci√≥n espec√≠fica de Skimage.

```python
from skimage.filters import threshold_otsu
```

```python
print("Umbral: ")
print(threshold_otsu(imgHomo))
```

Con este valor, ya sabemos que umbral aplicarle a la imagen para poder obtener la m√°scara.

```python
th_0 = 57
mask = imgHomo > th_0
img_seg = imgHomo*mask
plt.figure(figsize=(8, 8))
plt.subplot(121)
plt.imshow(mask, cmap='gray')
plt.title('M√°scara')
plt.subplot(122)
plt.imshow(img_seg, cmap='gray')
plt.title('Imagen Homog√©nea Transformada')
```

Como se puede apreciar con esta funci√≥n hemos conseguido extraer los granos de arroz en su totalidad quedando un resultado bastante satisfactorio.
